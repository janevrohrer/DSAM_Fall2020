```python
all_txt_files =[]
for root, dirs, files in os.walk(mydir):
    for file in files:
        if file.endswith(".txt"):
            all_txt_files.append(os.path.join(root, file))
n_files = len(all_txt_files)
all_txt_files [1]
```

```python
from natsort import natsorted
```

```python
all_files = natsorted(all_txt_files)
print(all_files)
```

```python
all_text = []
for i in all_files:
    with open(i) as f:
        txt = f.read()
    all_text.append(txt)
print (all_text)
```

```python
import spacy
from spacy.lang.en import English
nlp = English()

def text_cleanup(all_text):
    x = ['SPACE', 'SYM', 'NUM']
    doc = nlp(all_text.lower())
    processed = [i for i in doc if i.pos_ not in x]
    processed = [i.text for i in processed]
    return processed
```

```python
all_docs = [text_cleanup(i) for i in all_text]
print(all_docs)
```
