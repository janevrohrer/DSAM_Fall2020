```python
mydir ="/Users/janerohrer/Documents/eigner_scans/all_poems/all"
```

```python
all_txt_files =[]
for root, dirs, files in os.walk(mydir):
    for file in files:
        if file.endswith(".txt"):
            all_txt_files.append(os.path.join(root, file))
n_files = len(all_txt_files)
all_txt_files [1]
```

```python
import natsort
```

```python
from natsort import natsorted
```

```python
all_files = natsorted(all_txt_files)
```

```python
all_text = []
for i in all_files:
    with open(i) as f:
        txt = f.read()
    all_text.append(txt)
```

```python
import spacy
from spacy.lang.en import English
nlp = spacy.load("en_core_web_sm")

def text_cleanup(all_text):
    x = ['PUNCT', 'SPACE','SYM', 'NUM']
    doc = nlp(all_text.lower())
    processed = [i for i in doc if i.pos_ not in x]
    processed = [i.text for i in processed]
    return processed
```

```python
all_docs = [text_cleanup(i) for i in all_text]
print(all_docs)
```

```python
from collections import Counter

all_walks = []
for i in all_docs:
    freq = Counter(i)
    all_walks.append(freq['walks'])

print(all_walks)
```

```python
with open('output_dsam.txt', 'w') as f:
    for _list in all_docs:
        for _string in _list:
            f.write(str(_string) + '], [')
```
